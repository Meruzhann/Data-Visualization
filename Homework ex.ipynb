## Part 1: Data Cleaning and Exploration (Python & R)

1. Load the dataset. Check the first 5 rows.
2. Identify columns with missing values and their respective counts. Drop columns where more than
50% of the data is missing (store this version as a new dataset).
3. Convert the DATE OCC column to a datetime format. Extract the year, month, and day into separate
columns. Create a new column for the hour using the TIME OCC column.
4. Filter the dataset for crimes that occurred in 2023. Further filter crimes with the description
BURGLARY in the Crm Cd Desc column.
5. Group the data by AREA NAME and calculate the total number of crimes and the average victim age.
Sort the results by total crimes in descending order

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
crime_data = pd.read_csv('crime_data.csv')
crime_data.head(5)
```

```python
null_data = crime_data.isnull()
true_counts = null_data.apply(lambda col: (col == True).sum())
true_counts[true_counts!=0]
```

```python
crime_data.dropna(thresh = len(crime_data.index)//2, axis = 1)
```

```python
crime_data['DATE OCC'] = pd.to_datetime(crime_data['DATE OCC'], format='%m/%d/%Y %H:%M:%S AM')
crime_data['Year'] = crime_data['DATE OCC'].dt.year
crime_data['Month'] = crime_data['DATE OCC'].dt.month
crime_data['Day'] = crime_data['DATE OCC'].dt.day
crime_data['Hour'] = crime_data['TIME OCC'].apply(lambda x: int(str(x)[:2]))
crime_data
```

```python
crime_data[crime_data['Year'] == 2023]
```

```python
crime_data[(crime_data['Year'] == 2023) & (crime_data['Crm Cd Desc'] == 'BURGLARY')]
```

```python
crime_data.groupby("AREA NAME").agg(
    Average_Victim_Age=('Vict Age', 'mean'),
    Count=('Vict Age', 'size')
).sort_values(by='Count', ascending=False)
```
